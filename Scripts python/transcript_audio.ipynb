Objectif : transcrire les extraits audios en texte

```python
import whisper
import os

model = whisper.load_model("base")  # petit modèle rapide, tu peux choisir "small", "medium", "large"
input_folder = "../Audios/Splited_files"
output_folder = "../Transcriptions"

for filename in os.listdir(input_folder):
    if filename.endswith(".mp3"):
        audio_path = os.path.join(input_folder, filename)
        transcription_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.txt")

        result = model.transcribe(audio_path)
        text = result["text"]

        with open(transcription_path, "w", encoding="utf-8") as f:
            f.write(text)

        print(f"{filename} transcrit : {text}")
```
