Objectif : découper les pistes audio MP3 en extraits de 10 secondes

```python
import os
import subprocess
import random

# Dossier contenant les fichiers audio
input_folder = "../Audios/Original_files"
# Dossier où sauvegarder les extraits
output_folder = "../Audios/Splited_files"

clip_min_sec = 10
clip_max_sec = 15

for filename in os.listdir(input_folder):
    if filename.endswith(".mp3"):
        input_path = os.path.join(input_folder, filename)
        
        # Obtenir la durée totale en secondes
        result = subprocess.run(
            ["ffprobe", "-i", input_path, "-show_entries", "format=duration",
             "-v", "quiet", "-of", "csv=p=0"],
            capture_output=True, text=True
        )
        total_duration = float(result.stdout.strip())
        
        start = 0
        part = 1
        while start < total_duration:
            clip_duration = random.randint(clip_min_sec, clip_max_sec)
            end = min(start + clip_duration, total_duration)
            
            output_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}_part{part}.mp3")
            
            subprocess.run([
                "ffmpeg", "-y", "-i", input_path,
                "-ss", str(start), "-to", str(end),
                "-c", "copy", output_path
            ])
            
            start += clip_duration
            part += 1

        print(f"{filename} découpé en {part-1} extraits")


```
